```python
import polars as pl
from pathlib import Path
data_dir =  Path("/Users/jimmy/Code/FantasyFootball/data/ingest/fbref/advanced_match_stats")

df_keys = []
for player_file in data_dir.rglob("player/*/*.csv"):
    df = pl.read_csv(player_file).columns
    stat = str(player_file).split("/")[-2]
    name = player_file.name
    for col in df:
        df_keys.append({"name": name, "stat": stat, "col": col})
```

```python
df = pl.DataFrame(df_keys)
```
