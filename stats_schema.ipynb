```python
from pathlib import Path
import shutil

data_dir = Path("data/ingest/")
for f in data_dir.glob("**/*.csv"):
    df = pl.read_csv(f, ignore_errors=True)
    if "team" in str(f) and "Player" in df.columns and "keeper" not in str(f):
        new_file = Path(str(f).replace("ingest", "temp"))
        player_file = Path(str(f).replace("team", "player"))
        new_player_file = Path(
            str(f).replace("ingest", "temp").replace("team", "player")
        )
        if not new_file.parent.exists():
            new_file.parent.mkdir(parents=True, exist_ok=True)
        if not new_player_file.parent.exists():
            new_player_file.parent.mkdir(parents=True, exist_ok=True)
        shutil.move(f, new_file)
        shutil.move(player_file, new_player_file)
        print(f"Moved {f} to {new_file}")
```
