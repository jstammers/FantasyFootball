```python
from collections import defaultdict
stat_dict = defaultdict(list)
player_stats = Path("../data/FBRef/big5").glob("*_player.csv")
for p in player_stats:
	file_name = p.stem
	year = file_name.split("_")[0]
	stat_type = file_name[7:-7]
	df = pd.read_csv(p)
	stat_dict[stat_type].append(df)
	
stat_dict = {k: pd.concat(v) for k, v in stat_dict.items()}
```

```python
def plot_goals(df, player):
	df.query('Player == @player').sort_values('Season_End_Year').sort_values('Season_End_Year').plot(x='Season_End_Year', 
                                                            y=['Gls','Ast', 'G_minus_PK','Gls_Per'], subplots=True, layout=(2,2), figsize=(10, 5),title=f'{player} Stats')
```

```python
plot_goals(stat_dict['standard'], 'Lionel Messi') 
```

```python
plot_goals(stat_dict['standard'], 'Cristiano Ronaldo')
```

```python
plot_goals(stat_dict['standard'], 'Robert Lewandowski')
```
